### Limitation of Perceptron

The perceptron **cannot classify non-linearly separable data**. For example, it cannot solve problems like the XOR problem, where data classes cannot be separated by a straight line. This limitation arises because the perceptronâ€™s decision boundary is always linear. To handle non-linear data, more complex models like multi-layer perceptrons (MLPs) with hidden layers or other advanced neural networks are needed.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
or_data = pd.DataFrame()
and_data = pd.DataFrame()
xor_data = pd.DataFrame()
```

```python
or_data['X1'] = [1,1,0,0]
or_data['X2'] = [1,0,1,0]
or_data['Y'] = [1,1,1,0]
```

```python
and_data['X1'] = [1,1,0,0]
and_data['X2'] = [1,0,1,0]
and_data['Y'] = [1,0,0,0]
```

```python
xor_data['X1'] = [1,1,0,0]
xor_data['X2'] = [1,0,1,0]
xor_data['Y'] = [0,1,1,0]
```

```python
plt.figure(figsize=(10,7))
plt.scatter(or_data['X1'], or_data['X2'], c=or_data['Y'], cmap='viridis',s=400)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('OR Data')
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(and_data['X1'], and_data['X2'], c=and_data['Y'], cmap='viridis',s=400)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('AND Data')
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(xor_data['X1'], xor_data['X2'], c=xor_data['Y'], cmap='viridis',s=400)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('OR Data')
plt.show()
```

```python
from sklearn.linear_model import Perceptron

or_model = Perceptron()
and_model = Perceptron()
xor_model = Perceptron()

or_model.fit(or_data.iloc[:,0:2].values, or_data.iloc[:,-1].values)
and_model.fit(and_data.iloc[:,0:2].values, and_data.iloc[:,-1].values)
xor_model.fit(xor_data.iloc[:,0:2].values, xor_data.iloc[:,-1].values)

print(or_model.coef_,or_model.intercept_)
print(and_model.coef_,and_model.intercept_)
print(xor_model.coef_,xor_model.intercept_)
```

```python
x = np.linspace(-1,1,5)
y = - x + (1/2)
plt.figure(figsize=(10,7))
plt.scatter(or_data['X1'], or_data['X2'], c=or_data['Y'], cmap='viridis',s=400)
plt.plot(x,y)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('OR Data')
plt.show()
```

```python
x = np.linspace(-1,1,5)
y = - x + 1
plt.figure(figsize=(10,7))
plt.scatter(and_data['X1'], and_data['X2'], c=and_data['Y'], cmap='viridis',s=200)
plt.plot(x,y)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('AND Data')
plt.show()
```
