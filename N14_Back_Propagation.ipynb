### What is Backpropagation?
Backpropagation is an algorithm which trains the Neral Network. 
For a given data Backpropagation gives the optimal values of weights and biases by tracking the neural network backward for which the neural network will give more efficient predictions. 



Backpropagation: The Idea
- Goal: Adjust *weights* and *biases* to reduce error (loss).
- Works by *propagating error backward* through the network.
- Uses the *chain rule of differentiation* to calculate how a small change in a weight affects the overall loss.

## Steps:
1. *Initialize weights & biases.*
2. *Predict-Forward Propagation*:
   - Compute output using dot product of inputs and weights.
3. *Choose and Compute Loss* (e.g., MSE).
4. *Backpropagation*:
   - Compute gradients using chain rule.
   - Update weights and biases with gradient descent.
5. Repeat until error is minimized.
