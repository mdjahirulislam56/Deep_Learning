```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X, y  = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_clusters_per_class=1, random_state=42,class_sep=1.5,hypercube=False)
df = pd.DataFrame(X, columns=['Feature', 'X2'])
```

```python
y
```

```python
X.shape, y.shape
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(df['Feature'], df['X2'], c=y, cmap='bwr', alpha=0.7)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Synthetic Binary Classification Data')
plt.show()
```

```python
from sklearn.linear_model import Perceptron
P = Perceptron()
P.fit(X, y)
P.score(X, y)
```

```python
P.coef_, P.intercept_
```

```python
w1 = P.coef_[0][0]
w2 = P.coef_[0][1]
b = -(P.intercept_[0]/w2)
x= df['Feature']
m= -w1/w2
y_pred_line = m*x + b


plt.figure(figsize=(10, 7))
plt.plot(x , y_pred_line , 'r-',label='Decision Boundary')
plt.scatter(df['Feature'], df['X2'], c=y, cmap='bwr', alpha=0.7)
plt.legend()
plt.xlabel('Feature 1') 
plt.ylabel('Feature 2')
plt.title('Perceptron Decision Boundary')
plt.show()
```

```python
from mlxtend.plotting import plot_decision_regions #alternative to above plot
plt.figure(figsize=(10, 7))
plot_decision_regions(X, y, clf=P)
```
